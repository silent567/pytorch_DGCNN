```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from util import *
```

```python
N,C = 30,3
comm = np.zeros([N,C])
#prob = np.random.rand(C)
prob = [0.1,0.3,0.5]
print(prob)
comm[np.arange(N),np.random.choice(C,N)] = 1
graph = agm_comm2graph(comm,prob)
```

```python
index = np.argsort(np.argmax(comm,axis=1))
sgraph = graph[index][:,index]
plot_graph(sgraph,np.argmax(comm,axis=1)[index])
plt.show()
```

```python
g = nx.from_numpy_matrix(graph)
for i,c in enumerate(np.argmax(comm,axis=1)):
    g.nodes[i]['comm'] = c
print(g.nodes(data=True))
```
